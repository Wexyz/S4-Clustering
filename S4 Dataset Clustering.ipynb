## Unsupervised: Flat Clustering Analysis of the S4 Dataset

```python
from sklearn.datasets import load_files
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
file = "s4.txt"
#txt as Dataframe (df)
df = np.loadtxt(file)
```

```python
x = df[:1000,0]
y = df[:1000,1]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
xList = np.array([[0,0]])
```

```python
# making list out of x and y coordinates
for i in range(len(x)):
    temp = np.array([
        [int(x[i]),int(y[i])]
    ])
    xList = np.concatenate((xList,temp),axis=0)
```

```python
xList = np.delete(xList,0,0)
print(xList.shape)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=15)
km.fit(xList)
```

```python
centroids = km.cluster_centers_
```

```python
plt.scatter(x,y,s=10)
plt.scatter(centroids[:,0],centroids[:,1],marker="x",s=100,linewidth=3,zorder=10)
plt.show()
```
